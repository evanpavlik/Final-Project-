# The Best Neighborhood in Pittsburgh

There are many ways to define what makes a neighborhood good, but we focused mainly on **safety**. To us, a safe neighborhood is one with:

- Fewer arrests (less crime)
- Lower traffic levels (safer streets, fewer accidents, and easier parking)
- A strong amount of available services and resources for residents

By looking at these three different datasets, we tried to see which neighborhoods consistently performed well. Each dataset gives us a different angle on safety and quality of life.

Below are the visualizations we made for each metric, along with explanations of what they mean.


```python
adf = pd.read_csv("ArrestCountData.csv")

# I started by removing the rows that do not have a neighborhood listed because 
# those entries do not tell us where the arrest actually happened.
# Keeping them would give us inaccurate counts for each neighborhood.
clean_arrests = adf.dropna(subset=["INCIDENTNEIGHBORHOOD"])

# I counted how many arrests happened in each neighborhood.
# This lets me see which neighborhoods have more or fewer arrests overall.
arrest_counts = clean_arrests["INCIDENTNEIGHBORHOOD"].value_counts()

plt.figure(figsize=(15,5))

# I plotted the number of arrests for each neighborhood as a bar graph with a title, x and y axis, etc.
arrest_counts.plot(kind="bar")
plt.title("Amount of Arrests per Neighborhood")
plt.xlabel("Neighborhood")
plt.ylabel("Amount of Arrests")
plt.tight_layout()
plt.show()
```

### Interpreting the Arrest Data

This graph shows how many arrests occurred in each Pittsburgh neighborhood. The neighborhoods on the **left** have the highest number of arrests, while the neighborhoods on the **right** have the fewest.

Since we are measuring safety, we want neighborhoods with **lower arrest numbers**. High arrest counts can mean more crime or more police activity in that area.

From the graph, we can see that neighborhoods like the Central Business District, South Side Flats, and East Allegheny have very high arrest levels. On the opposite end, neighborhoods such as **Allegheny West**, Mt. Oliver Neighborhood, and North Shore have some of the **lowest arrest counts** in the entire dataset.

This suggests that **Allegheny West is one of the safer neighborhoods** in Pittsburgh based on arrests alone.


```python
tdf = pd.read_csv("TrafficCountData.csv")
cleaned = tdf.dropna(subset = ["average_daily_car_traffic"])
cleaner = cleaned.drop_duplicates(subset = ["neighborhood"])
cleanest = cleaner[["average_daily_car_traffic","neighborhood"]]
superclean = cleanest.sort_values(by="average_daily_car_traffic", ascending=True)

plt.figure(figsize=(30,8))
superclean.plot(kind="bar", x="neighborhood", y="average_daily_car_traffic", linewidth=1)
plt.show()
```

### Interpreting the Traffic Data

This graph shows the average daily car traffic in each neighborhood of Pittsburgh. The neighborhoods on the **left** have lighter traffic, while the ones on the **right** have very heavy traffic.

Lower traffic usually means:
- Safer roads for pedestrians and drivers  
- Fewer accidents  
- Less noise  
- Easier and less stressful parking

Higher traiffic could mean
- Unsafe roads for pedestrians and drivers
- More noise
- Harder or in worse case not able to park

Looking at the results, Allegheny West appears on the **low-traffic end** of the graph. This means it experiences far less congestion compared to many other areas in Pittsburgh.

This is another positive sign for safety and quality of life in Allegheny West.


```python
sdf= pd.read_csv("FacilityMetrics.csv")
newsdf = sdf[~sdf['report_group'].str.contains("Trends", na=False)]
counted = newsdf["geo_area_name"].value_counts()
plt.figure(figsize=(15,5))
counted.plot(kind="bar")
plt.title("Ammount of services per Neighborhood")
plt.xlabel("Neighborhood")
plt.ylabel("Ammount of services")
plt.tight_layout()
plt.show()
```

### Interpreting the Services Data

This graph shows the number of available city services or resources in each neighborhood. Almost all neighborhoods have between 55 and 70 services, which means most areas are supported fairly equally.

The main thing we want to check here is whether a safe and quiet neighborhood is missing important services. If a neighborhood has low arrests and light traffic, but barely any services, that would be a problem.

In this case, **Allegheny West falls within the normal range of services**. It is not missing resources, and its residents still have access to the same support systems that most other neighborhoods receive.

So while Allegheny West is quiet and safe, it still has enough services to support the people who live there.
