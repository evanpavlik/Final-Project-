# Real Estate Analysis
My second data set. I want to find the average sale price for each neighborhood to measure affordability.



## Load Dataset


```python
import pandas as pd

realestate = pd.read_csv("4ff5eb17-e2ad-4818-97c4-8f91fc6b6396.csv")
realestate.head()
```

## Data Cleaning


```python
realestate_clean = realestate[["neighborhood_name", "sale_price"]].dropna()
realestate_clean.head()
```

## Average Sale Price by Neighborhood


```python
avg_price = realestate_clean.groupby("neighborhood_name")["sale_price"].mean().sort_values()
avg_price
```

## Bar Chart 


```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
plt.barh(avg_price.index, avg_price.values)
plt.xlabel("Average Sale Price ($)")
plt.ylabel("Neighborhood")
plt.title("Average Home Sale Price by Neighborhood")
plt.tight_layout()
plt.show()
```



- Neighborhoods with lower average sale prices are more affordable.  

## Conclusion

After looking at the average sale prices by neighborhood, the results show clear differences in affordability across Pittsburgh. 
The neighborhoods with the lowest average prices are the most affordable, so they would score higher in this part of the bestness metric.
