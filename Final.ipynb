```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/zanderhodgson/Downloads/e03a89dd-134a-4ee8-a2bd-62c40aeebc6f.csv")
df = df.dropna(subset=["INCIDENTNEIGHBORHOOD"])
```

```python
counts = df["INCIDhoods by the number of arrests.ENTNEIGHBORHOOD"].value_counts()
plt.figure(figsize=(14,6))
counts.sort_values(ascending=False).plot(kind="bar")
plt.title("Number of Arrests per Pittsburgh Neighborhood")
plt.xlabel("Neighborhood")
plt.ylabel("Arrest Count")
plt.tight_layout()
plt.show()
```

```python
sdf= pd.read_csv("/Users/zanderhodgson/Downloads/33490823-a68e-40ce-b22d-cd7e03f0f2b2.csv")
newsdf = sdf[~sdf['report_group'].str.contains("Trends", na=False)]
counted = newsdf["geo_area_name"].value_counts()
plt.figure(figsize=(15,5))
counted.plot(kind="bar")
plt.title("Ammount of services per Neighborhood")
plt.xlabel("Neighborhood")
plt.ylabel("Ammount of services")
plt.tight_layout()
plt.show()
```

```python
tdf = pd.read_csv("/Users/zanderhodgson/Downloads/6dfd4f8f-cbf5-4917-a5eb-fd07f4403167.csv")
cleaned = tdf.dropna(subset = ["average_daily_car_traffic"])
cleaner = cleaned.drop_duplicates(subset = ["neighborhood"])
cleanest = cleaner[["average_daily_car_traffic","neighborhood"]]
superclean = cleanest.sort_values(by="average_daily_car_traffic",ascending=True)
plt.figure(figsize=(30,8))
superclean.plot(kind="bar", x="neighborhood",y ="average_daily_car_traffic", linewidth = 1)
```
